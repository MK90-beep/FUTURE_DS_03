```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()
df["VADER_Sentiment"] = df["Questions"].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df["VADER_Sentiment"]
```

```python
def vader_sentiment_category(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df["VADER_Label"] = df["VADER_Sentiment"].apply(vader_sentiment_category)
df["VADER_Label"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate all weightages
rating_distribution = df[['Weightage 1','Weightage 2','Weightage 3','Weightage 4','Weightage 5']].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=rating_distribution.index, y=rating_distribution.values, palette="viridis")
plt.title("Overall Rating Distribution (All Questions)")
plt.ylabel("Number of Responses")
plt.xlabel("Rating")
plt.show()
```

```python
# Group by Basic Course
dept_ratings = df.groupby("Basic Course")[['Weightage 1','Weightage 2','Weightage 3','Weightage 4','Weightage 5']].sum()

plt.figure(figsize=(12,6))
sns.heatmap(dept_ratings, annot=True, cmap="YlGnBu", fmt="d", linewidths=0.5)
plt.title("Ratings Distribution by Department (Basic Course)")
plt.ylabel("Department")
plt.xlabel("Rating")
plt.show()
```

```python
# Convert ratings into sentiment counts
df['Positive'] = df['Weightage 4'] + df['Weightage 5']
df['Neutral']  = df['Weightage 3']
df['Negative'] = df['Weightage 1'] + df['Weightage 2']

# Aggregate by department
dept_sentiments = df.groupby("Basic Course")[['Positive','Neutral','Negative']].sum()

# Normalize to percentages (for fair comparison)
dept_sentiments_pct = dept_sentiments.div(dept_sentiments.sum(axis=1), axis=0) * 100

# --- Visualization ---
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.heatmap(dept_sentiments_pct, annot=True, cmap="RdYlGn", fmt=".1f", linewidths=0.5)
plt.title("Sentiment Distribution by Department (Ratings-based)")
plt.ylabel("Department")
plt.xlabel("Sentiment Category")
plt.show()

# Overall sentiment pie chart
overall = dept_sentiments.sum()
overall.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6), colors=["blue", "green", "orange"])
plt.title("Overall Sentiment Distribution (Based on Ratings)")
plt.ylabel("")
plt.show()
```

```python
df["Average_Score"] = df["Average/ Percentage"].str.split("/").str[0].astype(float)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# 1. Average Score by Question
# ==============================
plt.figure(figsize=(12,6))
sns.barplot(x="Average_Score", y="Questions", data=df, palette="Set1")
plt.title("Average Score per Question")
plt.xlabel("Average Score (1–5)")
plt.ylabel("Survey Questions")
plt.show()

# ==============================
# 2. Overall Sentiment Pie Chart
# ==============================
overall = df[['Positive','Neutral','Negative']].sum()
overall.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6),
             colors=["green","gold","red"])
plt.title("Overall Sentiment Distribution")
plt.ylabel("")
plt.show()
# ==============================
# 3. Sentiment by Department (Stacked Bar)
# ==============================
dept_sentiments = df.groupby("Basic Course")[['Positive','Neutral','Negative']].sum()

dept_sentiments.plot(kind="bar", stacked=True, figsize=(12,6),
                     color=["#66c2a5","#fc8d62","#8da0cb"])  # teal, orange, purple

plt.title("Sentiment Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Number of Responses")
plt.legend(title="Sentiment")
plt.show()

# ==============================
# 4. Trend of Average Scores per Course
# ==============================
plt.figure(figsize=(10,6))
sns.boxplot(x="Basic Course", y="Average_Score", 
            data=df, palette="coolwarm")  # gradient blue–red

plt.title("Distribution of Average Scores by Department")
plt.xlabel("Department")
plt.ylabel("Average Score")
plt.show()

```
